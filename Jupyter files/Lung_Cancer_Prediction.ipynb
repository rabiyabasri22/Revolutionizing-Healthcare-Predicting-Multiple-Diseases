**LUNG CANCER DETECTION**

Importing Libraries

```python
import numpy as np
import pandas as pd
```

Data Collection and Data Analysis

```python
# Loading the dataset to pandas dataframe
lungcancer_ds=pd.read_csv('lungcancer.csv')
```

```python
# Information about the DataFrame
lungcancer_ds.info()
```

```python
# Print first 5 rows of dataset
lungcancer_ds.head()
```

```python
# Print last 5 rows of dataset
lungcancer_ds.tail()
```

```python
# Number of rows and columns in the dataset
lungcancer_ds.shape
```

```python
# Getting the statistical measures of the data (like mean,standard deviation etc..)
lungcancer_ds.describe()
```

```python
# Number of missing values in the datset
lungcancer_ds.isnull().sum()
```

Here

1 indicates has Lung Cancer and

0 indicates Doesn't Have Lung Cancer

```python
# Grouping data based on the target Variable
lungcancer_ds.groupby('LUNG_CANCER').mean()
```

Data Pre - Processing

```python
# Separating the data and labels
X=lungcancer_ds.drop(columns=['GENDER','LUNG_CANCER'],axis=1)
```

```python
Y=lungcancer_ds['LUNG_CANCER']
```

```python
print(X)
```

```python
print(Y)
```

Using Matplotlib for plotting Histograms

```python
import matplotlib.pyplot as plt
```

```python
lungcancer_ds.hist(figsize=(11,11))
plt.show()
```

Splitting the data into

Training data and

Testing data

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
# Checking the shape of X_train and X_test
print(X.shape, X_train.shape, X_test.shape)
```

Training the Model

```python
from sklearn.ensemble import RandomForestClassifier
```

We are making use of Random Forest Classifier


```python
clf = RandomForestClassifier(n_estimators=100, max_depth=10, random_state=42)
```

```python
#training the Random Forest Classifier model with Training data
clf.fit(X_train, Y_train)
```

Model Evaluation

Used to check how many times our model is predicting correctly

```python
from sklearn.metrics import accuracy_score
```

Calculating Accuracy Score

```python
# accuracy score on the training data
X_train_prediction = clf.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy score of the training data:',training_data_accuracy)
```

```python
# accuracy score on the testing data
X_test_prediction = clf.predict(X_test)
testing_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score of the testing data:',testing_data_accuracy)
```

Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
```

```python
print(f"Confusion Matrix =\n {confusion_matrix(Y_test,X_test_prediction)}")
```

Classification Report

```python
from sklearn.metrics import classification_report
```

```python
print(f"Classification Report =\n{classification_report(Y_test,X_test_prediction)}")
```

Building a Predictive System

```python
input_data = (75,1,2,1,1,2,2,2,2,1,2,2,1,1)

# change the input data to a numpy array
input_data_as_numpy_array= np.asarray(input_data)

# reshape the numpy array as we are predicting for only on instance
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

#std_data=scaler.transform(input_data_reshaped)

prediction = clf.predict(input_data_reshaped)
print(prediction)

if (prediction[0]== 0):
  print('The Person doesnot have Lung Cancer')
else:
  print('The Person have Lung Cancer')
```
